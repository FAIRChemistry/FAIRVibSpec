```python
import rich

import sif_parser
```

### Extract data and info from SIF file

```python
data, info = sif_parser.np_open("./data/Simon/298K-ADY79-10s-10acc.sif")

rich.print(f"Data array store in SIF file:\n{data}")
rich.print(f"Metadata, called info, stored besides data in SIF file:\n{info}")
```

### Extract calibration data from SIF info

```python
wavelengths = sif_parser.utils.extract_calibration(info)

rich.print(f"Calibration data as stored in SIF file metadata:\n{info["Calibration_data"]}")
rich.print(f"Actual wavelengths extracted from metadata:\n{wavelengths}")
```
